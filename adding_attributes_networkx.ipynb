```python
import networkx as nx
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("116thCongress.csv")
```

```python
df['score'] = df['score'].replace(np.nan, 0)
```

```python
G = nx.read_gml("following_graph.gml")
```

```python
for i in range(0,len(df)):
    x = df['twitter_handle'][i]
    y = df['twitter_handle_2'][i]
    z = df['twitter_handle_3'][i]
    name = df['name'][i]
    state = df['state'][i]
    party = df['party'][i]
    score = df['score'][i]
    group = df['congress_group'][i]
    mapping = {x:name}
    if not pd.isna(z):
        G = nx.contracted_nodes(G, x, y)
        G = nx.contracted_nodes(G, x, z)
        G = nx.relabel_nodes(G, mapping)
        print(name,":",x,y,z)
    elif not pd.isna(y):
        G = nx.contracted_nodes(G, x, y)
        G = nx.relabel_nodes(G, mapping)
        print(name,":",x,y)
    elif not pd.isna(x):
        G = nx.relabel_nodes(G, mapping)
        print(name,":",x)
    else:
        G.add_node(name)
    G.nodes[name].update({'state':state, 'party':party, 'score':score, 'house':group})
```

```python
deg = nx.degree_centrality(G)
cc = nx.closeness_centrality(G)
bet = nx.betweenness_centrality(G)
```

```python
nx.set_node_attributes(G, deg, "degree")
nx.set_node_attributes(G, bet, "betweenness")
nx.set_node_attributes(G, cc, "closeness")
```

```python
nx.write_gml(G, "bipartisanscoredgraph.gml")
```
