```python
import networkx as nx
import pandas as pd
```

```python
df = pd.read_csv("116thCongress.csv")
```

```python
df.head()
```

```python
G = nx.read_gml("following_graph.gml")
```

```python
twitter_handle_1 = list(df['twitter_handle'])
twitter_handle_2 = list(df['twitter_handle_2'])
twitter_handle_3 = list(df['twitter_handle_3'])
```

```python
unmergable = []
for node in G.nodes():
    if node not in twitter_handle_1:
        if node not in twitter_handle_2:
            if node not in twitter_handle_3:
                unmergable.append(node)
```

```python
print(unmergable)
```

```python
deg = nx.degree_centrality(G)
cc = nx.closeness_centrality(G)
bet = nx.betweenness_centrality(G)
```

```python
nx.set_node_attributes(G, deg, "degree")
nx.set_node_attributes(G, bet, "betweenness")
nx.set_node_attributes(G, cc, "closeness")
```

```python
df = pd.DataFrame.from_dict({
    'node': list(deg.keys()),
    'degree': list(deg.values()),
    'closeness': list(cc.values()),
    'betweenness': list(bet.values())    
})
df.head()
```

```python
df.sort_values('degree', ascending=False)
```

```python
df.sort_values('closeness', ascending=False)
```

```python
df.sort_values('betweenness', ascending=False)
```
